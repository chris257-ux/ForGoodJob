罗马数字包含以下七种字符: I， V， X， L，C，D 和 M。

字符          数值
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
例如， 罗马数字 2 写做 II ，即为两个并列的 1 。12 写做 XII ，即为 X + II 。 27 写做  XXVII, 即为 XX + V + II 。

通常情况下，罗马数字中小的数字在大的数字的右边。但也存在特例，例如 4 不写做 IIII，而是 IV。数字 1 在数字 5 的左边，所表示的数等于大数 5 减小数 1 得到的数值 4 。同样地，数字 9 表示为 IX。这个特殊的规则只适用于以下六种情况：

I 可以放在 V (5) 和 X (10) 的左边，来表示 4 和 9。
X 可以放在 L (50) 和 C (100) 的左边，来表示 40 和 90。 
C 可以放在 D (500) 和 M (1000) 的左边，来表示 400 和 900。
给定一个罗马数字，将其转换成整数。

 

示例 1:

输入: s = "III"
输出: 3
示例 2:

输入: s = "IV"
输出: 4
示例 3:

输入: s = "IX"
输出: 9
示例 4:

输入: s = "LVIII"
输出: 58
解释: L = 50, V= 5, III = 3.
示例 5:

输入: s = "MCMXCIV"
输出: 1994
解释: M = 1000, CM = 900, XC = 90, IV = 4.
 

提示：

1 <= s.length <= 15
s 仅含字符 ('I', 'V', 'X', 'L', 'C', 'D', 'M')
题目数据保证 s 是一个有效的罗马数字，且表示整数在范围 [1, 3999] 内
题目所给测试用例皆符合罗马数字书写规则，不会出现跨位等情况。
IL 和 IM 这样的例子并不符合题目要求，49 应该写作 XLIX，999 应该写作 CMXCIX 。
关于罗马数字的详尽书写规则，可以参考 罗马数字 - 百度百科。

```python
最优算法设计思路
我们可以采用遍历法：

使用一个字典存储罗马字符与对应数值的映射。

遍历字符串 s，从左到右依次处理每个字符。

如果当前字符的值小于下一个字符的值，则减去当前字符的值（表示特殊情况，如 IV 表示 4）。

否则，直接加上当前字符的值。

最终累加的结果即为转换后的整数。

这种方法的时间复杂度为 O(n)，其中 n 是字符串的长度。

时间复杂度与空间复杂度分析
时间复杂度：O(n)，只需遍历字符串一次。

空间复杂度：O(1)，使用了一个固定大小的字典。
```

```python
def romanToInt(s):
    """
    将罗马数字字符串转换为整数。
    
    :param s: str，罗马数字字符串
    :return: int，转换后的整数
    """
    # 罗马字符与对应数值的映射
    roman_to_int = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }
    
    total = 0  # 初始化结果
    n = len(s)
    
    # 遍历字符串
    for i in range(n):
        # 如果当前字符的值小于下一个字符的值，则减去当前字符的值
        if i < n - 1 and roman_to_int[s[i]] < roman_to_int[s[i + 1]]:
            total -= roman_to_int[s[i]]
        else:
            total += roman_to_int[s[i]]
    
    return total

# 测试用例 1
s = "III"
print("测试用例 1 输出:", romanToInt(s))  # 输出: 3

# 测试用例 2
s = "IV"
print("测试用例 2 输出:", romanToInt(s))  # 输出: 4

# 测试用例 3
s = "IX"
print("测试用例 3 输出:", romanToInt(s))  # 输出: 9

# 测试用例 4
s = "LVIII"
print("测试用例 4 输出:", romanToInt(s))  # 输出: 58

# 测试用例 5
s = "MCMXCIV"
print("测试用例 5 输出:", romanToInt(s))  # 输出: 1994
```
