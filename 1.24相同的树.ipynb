给你两棵二叉树的根节点 p 和 q ，编写一个函数来检验这两棵树是否相同。

如果两个树在结构上相同，并且节点具有相同的值，则认为它们是相同的。

 

示例 1：


输入：p = [1,2,3], q = [1,2,3]
输出：true
示例 2：


输入：p = [1,2], q = [1,null,2]
输出：false
示例 3：


输入：p = [1,2,1], q = [1,1,2]
输出：false
 

提示：

两棵树上的节点数目都在范围 [0, 100] 内
-104 <= Node.val <= 104

最优算法设计思路
使用 递归法 比较两棵树的节点：

若两棵树的当前节点均为空，返回 True。

若一棵树的当前节点为空而另一棵不为空，返回 False。

若当前节点的值不同，返回 False。

递归比较左子树和右子树。

时间复杂度与空间复杂度分析
时间复杂度：O(n)，其中 n 是二叉树的节点数。每个节点被访问一次。

空间复杂度：O(h)，其中 h 是二叉树的高度。递归调用栈的深度最多为 h。

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def isSameTree(p, q):
    if not p and not q:  # 两棵树均为空
        return True
    if not p or not q:  # 一棵树为空，另一棵不为空
        return False
    if p.val != q.val:  # 节点值不同
        return False
    return isSameTree(p.left, q.left) and isSameTree(p.right, q.right)  # 递归比较左右子树

# 辅助函数：根据列表构建二叉树
def build_tree(lst, index=0):
    if index >= len(lst) or lst[index] is None:
        return None
    root = TreeNode(lst[index])
    root.left = build_tree(lst, 2 * index + 1)
    root.right = build_tree(lst, 2 * index + 2)
    return root

# 测试用例 1
p1 = build_tree([1, 2, 3])
q1 = build_tree([1, 2, 3])
print("测试用例 1 输出:", isSameTree(p1, q1))  # 输出: True

# 测试用例 2
p2 = build_tree([1, 2])
q2 = build_tree([1, None, 2])
print("测试用例 2 输出:", isSameTree(p2, q2))  # 输出: False

# 测试用例 3
p3 = build_tree([1, 2, 1])
q3 = build_tree([1, 1, 2])
print("测试用例 3 输出:", isSameTree(p3, q3))  # 输出: False

# 测试用例 4
p4 = build_tree([])
q4 = build_tree([])
print("测试用例 4 输出:", isSameTree(p4, q4))  # 输出: True
```

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution(object):
    def isSameTree(self, p, q):
        """
        :type p: Optional[TreeNode]
        :type q: Optional[TreeNode]
        :rtype: bool
        """
        if not p and not q:
            return True
        elif not p or not q:
            return False
        elif p.val!=q.val:
            return False
        # else: return True
        left1=self.isSameTree(p.left,q.left)
        rig1=self.isSameTree(p.right,q.right)
        return left1 and rig1
        
```

注释部分 # else: return True，如果将其取消注释（即保留 else: return True），代码的逻辑会出错。原因在于这段代码的逻辑结构和递归调用的顺序。
为什么 else: return True 是错误的？
在你的代码中，else: return True 的位置是错误的。如果保留这一行，代码的逻辑会变成：
Python

elif p.val != q.val:

这意味着，只要 p.val == q.val，函数就会直接返回 True，而不会继续递归检查左右子树。这显然是错误的，因为即使根节点的值相同，左右子树可能仍然不同。
